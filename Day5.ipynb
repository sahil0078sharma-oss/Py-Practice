```python
# concept of constructor
class student:
  def __init__(self,name,roll_no): # constructor self(name of the constructor)
    self.name = name
    self.roll_no = roll_no

s = student("sahil",888)
print(s.name)
```

```python
class student:
  def __init__(self,name,roll_no,age):
    self.name = name
    self.roll_no = roll_no
    self.age = age

s = student("sahil",888,22)
print(s.name)
print(s.roll_no)
print(s.age)
```

```python
class student:
  def __init__(this,name = "sahil",roll_no = 888,age = None): # passed default parameters
    this.name = name
    this.roll_no = roll_no
    this.age = age

s = student("sahil",888)
print(s.name,s.roll_no,s.age)

s1 = student()
print(s1.name,s1.roll_no,s1.age)
```

```python
# example
class Student:
  def __init__(self,*args):
    self.data = args


s = Student({"users":["nikhil","raj","abhishek","vineet"]},{"adderess":{"home-town":"kukas","office":"jaipur"}})
print(
    *s.data[0]["users"],
    s.data[1]["adderess"]["home-town"],
    s.data[1]["adderess"]["office"]
)
```

```python
class Student:
  def __init__(self,*args):
    self.data = args


s = Student({"users":["nikhil","raj","abhishek","vineet"]},{"address":{"home-town":"kukas","office":"jaipur"}})
print() # all users , kukas ,jaipur

for i in s.data[0]['users']:
  print(i)


print(s.data[1]['address']['home-town'])
print(s.data[1]['address']['office'])
```
