# Operators


```python
# Arithmetic operator
a = 2
b = 5

sum = a+b
diff = a-b
mul = a*b
div = a/b

print(sum)
print(diff)
print(mul)
print(div)
print(a%b)   # modulo operator
print(a ** b) # power operator
```

```python
# Relational operator
a = 5
b = 7

print(a>b)
print(a == b) # is equals to
print(a != b) # not equals to

print(a >= b) # greater then equal to
print(a <= b) # less than equal to
```

```python
# Assignment operator
a = 10
b = 10
c = b + a # assigning values
print(c)

a = 10
a += 10 # a = 10 + 10
print(a)

a -= 10
print(a)
a *= 10
print(a)

a %= 10
print(a)

a **= 10
print(a)
```

```python
# Logical operators
print(not False)
print(not True)

a = True
b = False
print("The answer is =",a and b)

c = False
d = False
print("The answer is =", c or d  )

e = 10
f = 20
print("Result =",(e == f) and (f > e))
```
